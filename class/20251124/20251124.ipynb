```python
krx = "https://data.krx.co.kr/comm/bldAttendant/getJsonData.cmd"
```

```python
payload = {"bld":"dbms/MDC/STAT/standard/MDCSTAT01901",
"locale":"ko_KR",
"mktId":"ALL",
"share":"1",
"csvxls_isNo":"false"}
```

```python
requests.post(krx, data=payload)
```

```python
head = {'user-agent' :
"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36",
"origin" : "https://data.krx.co.kr",
'referer' : "https://data.krx.co.kr/contents/MDC/MDI/mdiLoader/index.cmd?menuId=MDC0201020201"}
r = requests.post(krx, data=payload, headers=head)
```

```python
# KOSPI만 네이버 주식 데이터 url 생성하기
# 20250101 ~ 20251124
url = "https://m.stock.naver.com/front-api/external/chart/domestic/info?symbol={}&requestType=1&startTime=20250101&endTime=20251124&timeframe=day"
for stock in r.json()['OutBlock_1']:
    if stock['MKT_TP_NM'] == "KOSPI":
        print(url.format(stock['ISU_SRT_CD']))
```
